## Задание 1

Сравним между собой попарно ряд нормально распределённых выборок с равным средним и сигмой без использования поправок.

```python
# Сгенерируем 4 выборки с равными средними и сигмой:
NUM_SAMPLES = 4
SAMPLE_SIZE = 10000
samples = []
for i in range(NUM_SAMPLES):
    samples.append(np.random.normal(0.0, 1.0, SAMPLE_SIZE))
```

```python
# Сравним выборки с помощью t-критерия:
alpha = 0.05
alpha_crit_left = alpha / 2
alpha_crit_right = 1 - alpha / 2
cnt = 0
for i in range(len(samples)):
    for j in range(i + 1, len(samples)):
        t, p = stats.ttest_ind(samples[i], samples[j])
        print(f'Проводим сравнение выборки {i} и {j}. Значение t-критерия {t}, p-value {p}')
        if p < alpha_crit_left or p > alpha_crit_right:
            cnt += 1
            print('Разница значима')
        else:
            print('Нет значимой разницы')
n_comparisons = len(samples) * (len(samples) - 1) / 2
print(f'Мы выявили значимую разницу в {cnt} из {n_comparisons} сравнений. Процент обнаружения значимой разницы {cnt/n_comparisons}')                 
```

## Задание 2

Сравним попарно между собой 20 выборок с неравными средними и сигмой, используя поправку Бонферрони.

```python
# Сгенерируем 20 выборок c неравными средними и сигмой:
NUM_SAMPLES = 20
SAMPLE_SIZE = 10000
samples = []
for i in range(NUM_SAMPLES):
    samples.append(np.random.normal(0.0 - i * 0.008, 1.0 + i * 0.01, SAMPLE_SIZE))
```

```python
# Определим новый уровень значимости с помощью поправки Бонферрони:
def Bonferroni(samples, alpha = 0.05):
    n_comparisons = len(samples) * (len(samples) - 1) / 2
    alpha /= n_comparisons
    return alpha
```

```python
# Сравним выборки с помощью t-критерия, учитывая новый уровень значимости:
alpha_bonf = Bonferroni(samples)
alpha_crit_left = alpha_bonf / 2
alpha_crit_right = 1 - alpha_bonf / 2
cnt_bonf = 0
for i in range(len(samples)):
    for j in range(i + 1, len(samples)):
        t, p = stats.ttest_ind(samples[i], samples[j])
        if p < alpha_crit_left or p > alpha_crit_right:
            cnt_bonf += 1
n_comparisons = len(samples) * (len(samples) - 1) / 2       
print(f'Мы выявили значимую разницу в {cnt_bonf} из {n_comparisons} сравнений. Процент обнаружения значимой разницы {cnt_bonf/n_comparisons}')                 
```

```python
# Теперь определим уровень значимости и посчитаем значения t-критерия с помощью поправки Холма-Бонферрони:
def Kholm_Bonferroni(samples, alpha = 0.05):
    p_values = []
    for i in range(len(samples)):
        for j in range(i + 1, len(samples)):
            t, p = stats.ttest_ind(samples[i], samples[j])
            p_values.append((min(p, 1 - p), i, j))
    p_values = sorted(p_values, key=lambda x: x[0])
    from pprint import pprint
    cnt = 0
    for idx, (p, i, j) in enumerate(p_values):
        p = min(p, 1 - p)
        if p < alpha  / 2 / (len(p_values) - (idx + 1) + 1):
            cnt += 1
    print(f'Мы выявили значимую разницу в {cnt} из {len(p_values)} сравнений. Процент обнаружения значимой разницы {cnt/len(p_values)}')
            
```

```python
Kholm_Bonferroni(samples)
```

Таким образом, мы убедились в том, что поправка Холма-Бонферрони действительно позволяет немного лучше, чем поправка Бонферрони, обнаружить значимую разницу и отклонить гипотезу H0, когда изменения есть.
